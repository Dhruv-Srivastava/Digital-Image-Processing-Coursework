### Dependencies

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Reading Images

```python
foreground = cv.imread("./assets/earth.jpg")
background = cv.imread("./assets/background.jpg")
```

```python
foreground = cv.cvtColor(foreground, cv.COLOR_BGR2RGB)
background = cv.cvtColor(background, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(foreground)
plt.figure()
plt.imshow(background)
```

```python
foreground.shape == background.shape
```

### Manipulate images

```python
foreground = cv.resize(foreground, (1000,1000))
background = cv.resize(background, (1000,1000))
```

```python
plt.imshow(foreground)
plt.figure()
plt.imshow(background)
```

### Image Addition

```python
img_add = cv.add(foreground,background)
plt.imshow(img_add)
```

```python
img_add_weighted = cv.addWeighted(foreground, 0.7, background, 0.3, 0)
plt.imshow(img_add_weighted)
```
