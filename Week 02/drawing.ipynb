### Dependencies

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Reading the image

```python
img = cv.imread("./assets/ManUnited.jpg")
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

### Drawing

#### 1) Line

```python
height,width,layers = img.shape
height,width,layers
```

```python
#Draw a green line from top left corner to bottom right corner

img = cv.line(img, (0,0), (width,height), (0,255,0), 2)
plt.imshow(img)
```

```python
# Draw a blue line from bottom left corner to top right corner
img = cv.line(img, (0,height), (width,0), (0,0,255), 2)
plt.imshow(img)
```

#### 2) Rectangle

```python
img = cv.rectangle(img, (160,15), (190,55), (255,0,0), 1)
plt.imshow(img)
```

#### 3) Circle

```python
img = cv.circle(img, (60,50), 30, (255,255,255), 1)
plt.imshow(img)
```

#### 4) Ellipse

```python
img = cv.ellipse(img, (115, 35), (25,35), 0, 0, 360, (255,255,255), 2)
plt.imshow(img)
```

#### 5) Text

```python
img = cv.putText(img, "Rashford",(135,12),cv.FONT_HERSHEY_SIMPLEX,0.4,(0,255,0,),1, cv.LINE_AA)
plt.imshow(img)
```

```python
img_resized = cv.resize(img,(200,100))
plt.imshow(img_resized)
```
