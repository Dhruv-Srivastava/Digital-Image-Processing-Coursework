```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.ones((10,10,3),np.uint8)*np.array([255,0,0],np.uint8)
plt.imshow(a)
```

```python
# Let's make a chess board, where each grid would be 10 x 10

grid = np.ones((10,10,3), np.uint8)*255
row  = grid

# Create first row

for _ in range(7):
    grid = 255 - grid
    row = np.concatenate((row,grid), axis = 1)
    
# Create remaining 7 rows
chess_board = row

for _ in range(7):
    row = 255 - row
    chess_board = np.concatenate((chess_board,row), axis = 0)

plt.imshow(chess_board)
    
    
```
